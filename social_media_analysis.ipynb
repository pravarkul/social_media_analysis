```python
import pandas as pd
import numpy as np
df=pd.read_csv("D:\working datasets\Social Media User Behavior\instagram_usage_lifestyle-RAW.csv")
df.head()
df.info()
df.describe(include='all')
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['last_login_date'] = pd.to_datetime(df['last_login_date'])
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[12,18,25,35,45,55,65],
    labels=['13-18','19-25','26-35','36-45','46-55','56-65']
)
```

```python
df['life_stage'] = np.where(
    (df['relationship_status'] == 'Married') | (df['has_children'] == 'Yes'),
    'Family',
    'Individual'
)
```

```python
df['daily_engagement_actions'] = (
    df['likes_given_per_day'] +
    df['comments_written_per_day'] +
    df['stories_viewed_per_day'] +
    df['reels_watched_per_day']
)
```

```python
df['weekly_dm_activity'] = (
    df['dms_sent_per_week'] +
    df['dms_received_per_week']
)
```

```python
df['total_time_spent_per_day'] = (
    df['time_on_feed_per_day'] +
    df['time_on_explore_per_day'] +
    df['time_on_messages_per_day'] +
    df['time_on_reels_per_day']
)
```

```python
df['daily_session_minutes'] = (
    df['average_session_length_minutes'] * df['sessions_per_day']
)
```

```python
df['user_type'] = np.where(
    df['posts_created_per_week'] > 3,
    'Creator',
    'Consumer'
)
```

```python
df['engagement_segment'] = pd.qcut(
    df['user_engagement_score'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Power']
)
```

```python
df['healthy_lifestyle'] = np.where(
    (df['exercise_hours_per_week'] >= 3) &
    (df['sleep_hours_per_night'] >= 7) &
    (df['diet_quality'].isin(['Good','Excellent'])),
    'Healthy',
    'Unhealthy'
)
```

```python
df.to_csv("D:\working datasets\Social Media User Behavior\social_media_user_analysis_clean.csv", index=False)
```
